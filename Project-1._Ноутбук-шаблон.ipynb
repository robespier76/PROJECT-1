# Преобразование данных

1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
#ваш код здесь

# Создаём функцию для преобразования значений в столбце
def get_education(education):
    # Согласно условию задачи, информация об образовании содержится в первых трёх словах и начинается с большой буквы
    education = education.split(sep=' ')
    education = education[0] + ' ' + education[1] + ' ' + education[2]
    if 'Неоконченное высшее' in education:
        return 'неоконченное высшее'
    elif 'Высшее' in education:
        return 'высшее'
    elif 'Среднее специальное' in education:
        return 'среднее специальное'
    elif 'Среднее' in education:
        return 'среднее'

# Создаём новый столбец, применяя функцию преобразования значений к столбцу "Образование и ВУЗ"
hh_df['Образование'] = hh_df['Образование и ВУЗ'].apply(get_education)
# Подсчитываем количество строк в датасете со значением "среднее в столбце "Образование"
mask_school = hh_df['Образование'] == 'среднее'
hh_df[mask_school].shape[0]
# удаляем столбец "Образование и ВУЗ" согласно условию задачи
hh_df.drop(['Образование и ВУЗ'], axis=1, inplace=True)
```

2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
#ваш код здесь

# Создаём функцию, вычленяющую пол и возраст из значения в столбце
def get_gender_age(info):
    info = info.split(',')
    age = int(info[1].strip()[:2])  # Учтем возможные пробелы и сделаем срез
    gender = 'М' if 'Мужчина' in info[0] else 'Ж'
    return pd.Series([gender, age], index=['Пол', 'Возраст'])

# Создаем два новых столбца сразу
hh_df[['Пол', 'Возраст']] = hh_df['Пол, возраст'].apply(get_gender_age)
# Выводим доли мужчин и женщин
hh_df['Пол'].value_counts(normalize=True)
# Выводим средний возраст соискателей
hh_df['Возраст'].mean()
# Удаляем первоначальный столбец с информацией
hh_df.drop(['Пол, возраст'], axis=1, inplace=True)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь

# создаём функцию для обработки информации об опыте работы
def get_experience(experience):
    # Если значение пустое или не указано, возвращаем NaN
    if type(experience) == float or (experience == 'Не указано'):
        return np.nan
    # Вычленяем из строки только информацию о длительности опыта
    experience = experience.split(' ')[2:6]
    # Вычисляем длительность опыта работы в месяцах
    words_years = ['год', 'года', 'лет']
    words_months = ['месяц', 'месяца', 'месяцев']
    months = 0
    for i, word in enumerate(experience):
        if word in words_years:
            months += int(experience[i-1]) * 12
            continue
        if word in words_months:
            months += int(experience[i-1])
            continue
    return months

# Создаём новый столбец с информацией об опыте работы в месяцах     
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_experience)
# Удаляем столбец со старой информацией
hh_df.drop(['Опыт работы'], axis=1, inplace=True)
# Вычисляем медианный опыт
hh_df['Опыт работы (месяц)'].median()
    
```

    * … , готов к командировкам , …

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь

# Создаём функцию для преобразования информации
def get_loc_info(info):
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']
    # Выделяем город из строки
    city = info[:info.find(' ')]
    # Преобразовываем информацию о городе согласно условию задачи
    if (city != 'Москва') and (city != 'Санкт-Петербург'):
        city = 'город-миллионник' if city in million_cities else 'другие'
    # Преобразовываем информацию о готовности к переезду согласно условию задачи
    if  ('не готов к переезду' in info) or ('не готова к переезду' in info) or ('не хочу переезжать' in info):
        relocate = False
    elif ('готов к переезду' in info) or ('готова к переезду' in info) or ('хочу переехать' in info):
        relocate = True
    else:
        relocate = False
        
    # Преобразовываем информацию о готовности к командировкам согласно условию задачи
    if ('не готов к командировкам' in info) or ('не готова к командировкам' in info):
        business_trip = False
    elif ('готов к командировкам' in info) or ('готова к командировкам' in info) or ('готов к редким командировкам' in info) or ('готова к редким командировкам' in info):
        business_trip = True
    else:
        business_trip = False
    # Возвращаем объект Series с данными для новых столбцов исходной таблицы
    return pd.Series([city, relocate, business_trip], index=['Город', 'Готовность к переезду', 'Готовность к командировкам'])

# Создаём новые столбцы в таблице путём преобразования информации о городе, готовности к переезду и командировкам
hh_df[['Город', 'Готовность к переезду', 'Готовность к командировкам']] = hh_df['Город, переезд, командировки'].apply(get_loc_info)
# Вычисляем процент соискателей из Санкт-Петербурга, а также процент тех, кто готов и к переезду, и к командировкам
mask1 = hh_df['Город'] == 'Санкт-Петербург'
mask2 = (hh_df['Готовность к переезду'] == True) & (hh_df['Готовность к командировкам'] == True)
print(hh_df[mask1].shape[0] / hh_df.shape[0] * 100)
print(hh_df[mask2].shape[0] / hh_df.shape[0] * 100)
# Удаляем из таблицы старый столбец «Город, переезд, командировки»    
hh_df.drop(['Город, переезд, командировки'], axis=1, inplace=True)    
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь

# Создаём функцию для обработки информации о занятости
def get_employment(info):
    # Создаём переменные для каждого вида занятости
    full_emp = True if 'полная занятость' in info else False
    partly_emp = True if 'частичная занятость' in info else False
    project_job = True if 'проектная работа' in info else False
    volunteering = True if 'волонтерство' in info else False
    traineeship = True if 'стажировка' in info else False
    # Возвращаем объект Series с данными для новых столбцов исходной таблицы
    return pd.Series([full_emp, partly_emp, project_job, volunteering, traineeship], index=['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка'])
    
# Создаём функцию для обработки информации о графике
def get_schedule(info):
    # Создаём переменные для каждого вида графиков работы
    full_day = True if 'полный день' in info else False
    remote_job = True if 'удаленная работа' in info else False
    shift_schedule  = True if 'сменный график' in info else False
    flex_schedule = True if 'гибкий график' in info else False
    shift_job = True if 'вахтовый метод' in info else False
    # Возвращаем объект Series с данными для новых столбцов исходной таблицы
    return pd.Series([full_day, remote_job, shift_schedule, flex_schedule, shift_job], index=['полный день', 'удаленная работа', 'сменный график', 'гибкий график', 'вахтовый метод'])
    
# Создаём новые столбцы в таблице путём преобразования информации о занятости и графике работы
hh_df[['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']] = hh_df['Занятость'].apply(get_employment)
hh_df[['полный день', 'удаленная работа', 'сменный график', 'гибкий график', 'вахтовый метод']] = hh_df['График'].apply(get_schedule)
# Вычисляем количество соискателей, согласных на проектную работу и волонтёрство 
mask1 = (hh_df['проектная работа'] == True) & (hh_df['волонтерство'] == True)
print(hh_df[mask1].shape[0])
# Вычисляем количество соискателей, согласных на гибкий график и ввхтовый метод работы
mask2 = (hh_df['вахтовый метод'] == True) & (hh_df['гибкий график'] == True)
print(hh_df[mask2].shape[0])
# Удаляем из исходной таблицы старые столбцы о занятости и графике
hh_df.drop(['Занятость'], axis=1, inplace=True) 
hh_df.drop(['График'], axis=1, inplace=True) 
```

6. (2 балла) Наконец, мы добрались до самого главного и самого важного - признака заработной платы **"ЗП"**. 
В чем наша беда? В том, что помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать, например:

Нам бы хотелось видеть заработную плату в единой валюте, например, в рублях. Возникает вопрос, а где взять курс валют по отношению к рублю?

На самом деле язык Python имеет в арсенале огромное количество возможностей получения данной информации, от обращения к API Центробанка, до использования специальных библиотек, например pycbrf. Однако, это не тема нашего проекта.

* "date" - дата,

Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь

# Функция возвращает сумму заработной платы и валюту, в которой она выражена
def get_curr(info):
    info = info.split(' ')
    # Сумма з/п
    sum = int(info[0])
    # Преобразуем наименование валюты з/п
    curr = info[1].replace('.', '')
    if curr == 'грн': curr = 'UAH'
    elif curr == 'белруб': curr = 'BYN'
    elif curr == 'сум': curr = 'UZS'
    # Возвращаем Series с суммой и наименованием валюты з/п
    return pd.Series([sum, curr], index=['Зарплата', 'Валюта зарплаты'])

# Преобразуем столбец 'Обновление резюме' к формату даты
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], dayfirst=True)
# Создаём столбцы с суммой и валютой з/п
hh_df[['Зарплата', 'Валюта зарплаты']] = hh_df['ЗП'].apply(get_curr)
# Получаем DF из файла с курсами валют
curr_df = pd.read_csv('data/ExchangeRates.csv')
# Удаляем из DF лишние столбцы
curr_df.drop(['per', 'time', 'vol'], axis=1, inplace=True)
# Преобразуем столбец 'date' к формату даты
curr_df['date'] = pd.to_datetime(curr_df['date'], dayfirst=True)
# Оставляем только информациб о дате в столбце 'Обновление резюме'
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], dayfirst=True).dt.date
# Так как в результате в столбце 'Обновление резюме' у нас получились данные типа object, снова преобразуем столбец в формат даты
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'])
# Присоединяем к DF с резюме данные по курсам валют, соединяем DF по двум столбцам - даты и валюты
hh_df = pd.merge(hh_df, curr_df, left_on=['Обновление резюме', 'Валюта зарплаты'], right_on=['date', 'currency'], how='left')
# Заполняем пропуски в столцах 'close' и 'proportion' в объединённом DF
hh_df['close'].fillna(value=1, inplace=True)
hh_df['proportion'].fillna(value=1, inplace=True)
# Вычисляем з/п в рублях и помещаем её в новый столбец
hh_df['ЗП (руб)'] = hh_df['Зарплата'] * hh_df['close'] / hh_df['proportion']
# Вычисляем медианную з/п
hh_df['ЗП (руб)'].median()
# Удаляем все лишние столбцы из DF
hh_df.drop(['ЗП', 'Зарплата', 'Валюта зарплаты', 'currency', 'date', 'close', 'proportion'], axis=1, inplace=True)
# Выводим информацию о DF для сверки
hh_df.info()
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# ваш код здесь

# Выделяем столбец с возрастом в отдельную структуру для последующего логарифмирования
log_distr_df = np.log(hh_df['Возраст'])
# Строим гистограмму распределения возраста
fig = px.histogram(data_frame=log_distr_df, x='Возраст', title='Логарифмическое распределение возраста соискателей')
# Строим вертикальные линии среднего, среднего минус 3 сигмы и среднего плюс 3 сигмы
fig.add_vline(log_distr_df.mean(), line_width=3)
fig.add_vline(log_distr_df.mean() + 3 * log_distr_df.std(), line_width=3, line_dash='dash')
fig.add_vline(log_distr_df.mean() - 3 * log_distr_df.std(), line_width=3, line_dash='dash')
fig.show("png")
fig.write_html("plotly/hist_logdistr_age_blowouts.html") # Записываем график в файл html для последующего отображения в GitHub

# Вычисляем показатель ассиметрии:
print(f'ассиметрия: {log_distr_df.skew()}') # Положительная, то есть правосторонняя

# Создаём функцию, вычисляющую выбросы по методу z-отклонений 
def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

# Получаем выбросы с помощью функции  
outliers, cleaned = outliers_z_score(hh_df, 'Возраст', log_scale=True)
print(f'выбросы по методу z-отклонений: {outliers.shape[0]}')
# Выводим таблицу выбросов
display(outliers.head())
# Удаляем выбросы
hh_df.drop(labels = outliers.index, inplace=True)
```
